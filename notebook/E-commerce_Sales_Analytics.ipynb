```python
import pandas as pd

df = pd.read_csv(r'F:\Project\E-commerce Sales Analytics\Data\raw\Superstore_Dataset.csv')
```

```python
df["Product Name"] = (df["Product Name"].str.slice(0, 35).where(df["Product Name"].str.len() <= 35,df["Product Name"].str.slice(0, 35) + "..."))
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y', errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y', errors='coerce')
```

```python
df["Postal Code"] = df["Postal Code"].fillna("05401")
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df['order_year'] = df['order_date'].dt.year
df['order_month'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()
df["quarter"] = df["order_date"].dt.quarter
df["shipping_days"] = (df["ship_date"] - df["order_date"]).dt.days
```

```python
import pandas as pd
from sqlalchemy import create_engine, text

HOST = "localhost"
PORT = 3306
USER = "root"
PASSWORD = "tushar"      
DATABASE = "ecsales"

df = pd.read_csv(r"F:\Project\E-commerce Sales Analytics\Data\processed\superstore_cleaned.csv")

print("CSV Loaded Successfully")
print(df.head())

engine = create_engine(
    f"mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}"
)

with engine.connect() as conn:
    conn.execute(text(f"CREATE DATABASE IF NOT EXISTS {DATABASE}"))
    conn.commit()

print(f"Database '{DATABASE}' is ready.")

engine = create_engine(
    f"mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DATABASE}"
)

df.to_sql(
    name="superstore",
    con=engine,
    if_exists="replace",   
    index=False
)

print("Table created successfully.")
print("Data inserted successfully.")

query = "SELECT * FROM superstore LIMIT 10"

result = pd.read_sql(query, engine)

print("\nFirst 10 Records:")
print(result)

print("\nProcess Completed Successfully.")
```
